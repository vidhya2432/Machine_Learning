# **Problem 1: Simple Linear Regression from Scratch**

**What is the goal?**

We want to find the best straight line that fits the given data points.

The line equation is:







# y=mx+b

**Why is this useful?**

It helps us predict values and understand the relationship between two variables.

```python
import numpy as np

def simple_linear_regression(x, y):
    """
    Computes slope (m) and intercept (b) using least squares.
    """
    x = np.array(x)
    y = np.array(y)

    x_mean = np.mean(x)
    y_mean = np.mean(y)

    m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
    b = y_mean - m * x_mean

    return round(m, 2), round(b, 2)

# Test case
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
print(simple_linear_regression(x, y))
```

# **Problem 2: R-Squared Calculation**

**What is R-Squared?**

R² tells us how good our model is.

Value close to 1 → very good model

Value close to 0 → poor model

Use the formula:

**R^2 = 1- SS res/ SS tot**

Calculate total variation in actual data.

Calculate error variation between actual and predicted values.

**Why is this useful?**

It helps us evaluate model accuracy.





```python
def r_squared(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

    return round(1 - (ss_res / ss_tot), 3)

# Test case
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
print(r_squared(y_true, y_pred))
```

# **Problem 3: Multiple Linear Regression**

**What is the goal?**

Predict output using multiple features instead of just one.



β=(X
T
 X)
 −1
  X
  T
   y



**Why is this useful?**

Real-world problems often depend on many variables.







```python
def multiple_linear_regression(X, y):
    X = np.array(X)
    y = np.array(y)

    beta = np.linalg.inv(X.T @ X) @ X.T @ y
    return beta.round(2).tolist()

# Test case
X = [[1, 1], [1, 2], [1, 3], [1, 4]]
y = [2, 4, 5, 4]
print(multiple_linear_regression(X, y))
```

# **Problem 4: Gradient Descent Optimization**


**What is Gradient Descent?**


It is an iterative method to find the best weights that minimize error.

Predict values











Calculate error

Update weights using gradients



**Why is this useful?**


Used when datasets are large and matrix inversion is slow.

```python
def gradient_descent(X, y, lr=0.01, iterations=100):
    X = np.c_[np.ones(len(X)), X]  # Add bias term
    y = np.array(y)
    weights = np.zeros(X.shape[1])
    loss_history = []

    for _ in range(iterations):
        predictions = X @ weights
        errors = predictions - y
        gradients = (2 / len(y)) * X.T @ errors
        weights -= lr * gradients
        loss = np.mean(errors ** 2)
        loss_history.append(loss)

    return weights.round(2).tolist(), loss_history

# Test case
X = [[1], [2], [3], [4]]
y = [2, 4, 6, 8]
weights, loss = gradient_descent(X, y)
print(weights)
```

# **Problem 5: Ridge Regression (L2 Regularization)**

**What problem does it solve?**

Prevents overfitting by penalizing large coefficients.

β=(X
T
 X+λI)
 −1
  X
  T
   y  ( regularization term (λI) to the normal equation.)

 **Why is this useful?**
   
 Improves model stability when features are correlated.

```python
def ridge_regression(X, y, lambda_):
    X = np.array(X)
    y = np.array(y)
    I = np.eye(X.shape[1])

    beta = np.linalg.inv(X.T @ X + lambda_ * I) @ X.T @ y
    return beta.round(2).tolist()

# Test case
X = [[1, 5], [1, 10], [1, 15]]
y = [3, 7, 11]
print(ridge_regression(X, y, 0.1))
```

# **Problem 6: Outlier Detection in Regression**

**What is an outlier?**

A data point that is far away from the regression line.



*  Calculate residuals (actual − predicted)

*  Find standard deviation of residuals


*   Mark points as outliers

      ∣ residual ∣  > threshold × std  


**Why is this useful?**


Outliers can mislead the model, so detecting them is important



```python
def detect_outliers(X, y, threshold=2.0):
    m, b = simple_linear_regression(X, y)
    y_pred = np.array(X) * m + b
    residuals = np.abs(np.array(y) - y_pred)

    std_dev = np.std(residuals)
    outliers = [i for i, r in enumerate(residuals) if r > threshold * std_dev]

    return outliers

# Test case
X = [1, 2, 3, 4, 5]
y = [2, 4, 6, 15, 10]
print(detect_outliers(X, y))
```

# **Problem 7: Polynomial Regression**

**What is Polynomial Regression?**

Used when data follows a curve, not a straight line.





**Why is this useful?**

Captures non-linear relationships in data.



```python
def polynomial_regression(x, y, degree):
    X = np.vander(x, degree + 1, increasing=True)
    y = np.array(y)

    beta = np.linalg.inv(X.T @ X) @ X.T @ y
    return beta.round(2).tolist()

# Test case
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
print(polynomial_regression(x, y, 2))
```
