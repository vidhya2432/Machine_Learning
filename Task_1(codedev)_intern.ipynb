```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Iris Dataset
df = pd.read_csv("/content/1) iris.csv")
```

### Task 1: Data Cleaning & Preprocessing

```python
print("First 5 Rows:")
display(df.head())

print("\nDataset Information:")
print(df.info())
```

```python
# Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Fill missing values if any
for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].mean())

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Remove Duplicates
print("\nDuplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# Standardize Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("\nData Cleaning Completed Successfully!")
```

### Task 2: Exploratory Data Analysis (EDA)

```python
print("\n===== SUMMARY STATISTICS =====")

print("\nMean:")
print(df.mean(numeric_only=True))

print("\nMedian:")
print(df.median(numeric_only=True))

print("\nMode:")
print(df.mode().iloc[0])

print("\nStandard Deviation:")
print(df.std(numeric_only=True))

print("\nDescribe:")
print(df.describe())
```

```python
# Correlation Matrix
plt.figure(figsize=(8,6))
sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Iris Dataset Correlation Matrix")
plt.show()
```

```python
# Histogram
df.hist(figsize=(10,8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
# Box Plots
plt.figure(figsize=(10,6))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title("Box Plot of Iris Features")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species'
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

### Task 3: Basic Data Visualization

```python
# Line Chart
plt.figure(figsize=(8,5))
plt.plot(df.index, df['sepal_length'])
plt.title("Sepal Length Line Chart")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.legend(["Sepal Length"])
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='sepal_width',
    y='petal_width',
    hue='species'
)
plt.title("Sepal Width vs Petal Width")
plt.show()
```

```python
# Save Cleaned Dataset
df.to_csv("cleaned_iris_dataset.csv", index=False)

print("\n===================================")
print("Task 1 Completed: Data Cleaning")
print("Task 2 Completed: EDA")
print("Task 3 Completed: Visualization")
print("Cleaned dataset saved as cleaned_iris_dataset.csv")
print("===================================")
```
